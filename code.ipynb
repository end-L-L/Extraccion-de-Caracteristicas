# Determinacion de Clases Para Datos Agrupados

```python
# Corroboramos los valores nulos
nulos = Micro_Retailer.isnull().sum()
nulos
```

```python
# calculamos el numero total de la poblacion "n"
Micro_Retailer['_latitude'].info()
n = 171
```

```python
# obtenemos el limite superior y el límite inferior de la columna objetivo
Max=Micro_Retailer['_latitude'].max()
Min=Micro_Retailer['_latitude'].min()
Limites = [Min, Max]
Limites
```

```python
# calculamos el rango r
R = Max - Min
R
```

```python
# calculamos el número de intervalos de clase "ni", aplocando la regla de Sturges
ni = 1+3.32*np.log10(n)
ni
```

```python
# calculamos el ancho del intervalo "i"
i = R/ni
i
```

# Creación de Categorias a partir de las clases

```python
# categorizacion de variables
#intervalos = np.linspace(Min, Max, ni + 1)
intervalos = np.linspace(18.9993095, 19.0752733, 9)
intervalos
```

```python
# creamos las categorias
categorias = ["Categoria 1", "Categoria 2", "Categoria 3", "Categoria 4", "Categoria 5", "Categoria 6", "Categoria 7", "Categoria 8"]
```

```python
# finalmente creamos las categorias en la columna numerica
Micro_Retailer['_latitude'] = pd.cut(Micro_Retailer['_latitude'], bins=intervalos, labels=categorias, right=False)
#Micro_Retailer['_latitude'] = pd.cut(x = Micro_Retailer['_latitude'], bins=intervalos, labels=categorias)
Micro_Retailer['_latitude']
```
